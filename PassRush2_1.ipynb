```python
import zipfile
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
!pip -q install dtaidistance
from dtaidistance import dtw

def safe_groupby_agg(df, group_cols, agg_spec, sort=False, observed=True):
    """
    Returns a normal DataFrame with group_cols as columns.
    Avoids pandas as_index=False insertion bug by using reset_index().
    Any agg whose source column is missing is dropped automatically.
    """
    keep = {}
    missing = []
    for out_name, (src_col, func) in agg_spec.items():
        if src_col in df.columns:
            keep[out_name] = (src_col, func)
        else:
            missing.append((out_name, src_col))
    if missing:
        print("⚠️ Dropping missing agg columns:", missing)

    g = df.groupby(group_cols, as_index=True, sort=sort, observed=observed).agg(**keep)
    return g.reset_index()

# =========================
# USER SETTINGS
# =========================
ZIP_PATH = "/content/nfl-big-data-bowl-2021 (2).zip"
WEEKS = range(1, 18)          # start small: range(1, 3) to debug faster
LOS_X_TOL = 2.0               # LOS proximity threshold in yards
K_FRAMES = 25                 # fixed-length representation (~2.5s at 10Hz)
CLUSTERS_K = 6                # try 4–10

PRESSURE_YARDS_IMMEDIATE = 2.0   # PRIMARY metric
PRESSURE_YARDS_POCKET    = 3.5   # secondary / context
FPS = 10.0
MAX_SECONDS = 4.0
MAX_FRAMES = int(MAX_SECONDS * FPS)


# =========================
# ZIP HELPERS (robust to subfolders)
# =========================
zf = zipfile.ZipFile(ZIP_PATH)

def find_in_zip(endswith_name: str) -> str:
    matches = [n for n in zf.namelist() if n.endswith(endswith_name)]
    if not matches:
        raise FileNotFoundError(f"Could not find file ending with: {endswith_name}")
    return matches[0]

def read_zip_csv(endswith_name: str, **kwargs) -> pd.DataFrame:
    path = find_in_zip(endswith_name)
    with zf.open(path) as f:
        return pd.read_csv(f, **kwargs)


# =========================
# LOAD BASE TABLES
# =========================
games   = read_zip_csv("games.csv")
plays   = read_zip_csv("plays.csv")
players = read_zip_csv("players.csv")

print("Loaded:", {"games": games.shape, "plays": plays.shape, "players": players.shape})

required_games = {"gameId", "homeTeamAbbr", "visitorTeamAbbr"}
required_plays = {"gameId", "playId", "possessionTeam", "playType", "typeDropback",
                  "numberOfPassRushers", "epa", "passResult"}

missing_games = sorted(list(required_games - set(games.columns)))
missing_plays = sorted(list(required_plays - set(plays.columns)))
if missing_games or missing_plays:
    raise ValueError({"missing_games": missing_games, "missing_plays": missing_plays})

games_small = games[["gameId", "homeTeamAbbr", "visitorTeamAbbr"]].copy()

# Keep player identity columns handy
players_small = players[["nflId", "displayName", "position"]].drop_duplicates("nflId").copy()


# =========================
# 1) FILTER PLAYS (CLEAN DROPBACK PASS)
# =========================
pass_plays = plays[plays["playType"].astype(str).str.contains("pass", case=False, na=False)].copy()
pass_plays = pass_plays[pass_plays["typeDropback"].eq("TRADITIONAL")].copy()

if "penaltyCodes" in pass_plays.columns:
    pass_plays = pass_plays[pass_plays["penaltyCodes"].isna()].copy()

pass_plays = pass_plays[pass_plays["passResult"].isin(["C", "I", "S"])].copy()
pass_plays = pass_plays[pass_plays["numberOfPassRushers"].between(3, 6)].copy()

pass_plays_small = pass_plays[
    ["gameId", "playId", "possessionTeam", "numberOfPassRushers", "epa", "passResult"]
].copy()

# Robustness guard (even though 2021 has epa)
if "epa" not in pass_plays_small.columns:
    pass_plays_small["epa"] = np.nan

print("Filtered pass plays:", len(pass_plays_small))


# =========================
# 2) WEEK ACCUMULATOR: LOAD ONE WEEK AT A TIME, APPEND INTO tp
# =========================
tp_all = []

for w in WEEKS:
    week_df = read_zip_csv(f"week{w}.csv")

    required_tracking = {"gameId", "playId", "frameId", "team", "event", "nflId", "x", "y"}
    missing_tracking = sorted(list(required_tracking - set(week_df.columns)))
    if missing_tracking:
        raise ValueError({"week": w, "missing_tracking_cols": missing_tracking})

    week_df["team"] = week_df["team"].astype(str).str.lower()
    week_df["week_loaded"] = w

    tp_week = week_df.merge(pass_plays_small, on=["gameId", "playId"], how="inner")
    tp_week = tp_week.merge(games_small, on="gameId", how="left")

    tp_all.append(tp_week)
    print(f"week{w}: added={len(tp_week):,} | accumulated={sum(len(x) for x in tp_all):,}")

tp = pd.concat(tp_all, ignore_index=True)

print("\nWeeks present in tp:")
print(tp["week_loaded"].value_counts().sort_index())
print("tp shape:", tp.shape)


# =========================
# 3) MAP TEAM ABBR + OFF/DEF TEAMS
# =========================
tp["teamAbbr"] = np.select(
    [tp["team"].eq("home"), tp["team"].eq("away"), tp["team"].eq("football")],
    [tp["homeTeamAbbr"], tp["visitorTeamAbbr"], "football"],
    default=np.nan
)

tp["offTeamAbbr"] = tp["possessionTeam"]
tp["defTeamAbbr"] = np.where(
    tp["offTeamAbbr"].eq(tp["homeTeamAbbr"]),
    tp["visitorTeamAbbr"],
    tp["homeTeamAbbr"]
)

# Ensure identity columns exist in tracking
if "displayName" not in tp.columns or "position" not in tp.columns:
    tp = tp.merge(players_small, on="nflId", how="left")


# =========================
# 4) FIND SNAP / END FRAME PER PLAY
# =========================
def first_frame(df, event_name):
    s = df.loc[df["event"].eq(event_name), "frameId"]
    return None if s.empty else int(s.iloc[0])

key_frames = (
    tp.groupby(["gameId", "playId"])
      .apply(lambda x: pd.Series({
          "snapFrame": first_frame(x, "ball_snap"),
          "passFrame": first_frame(x, "pass_forward"),
          "sackFrame": first_frame(x, "qb_sack"),
      }))
      .reset_index()
)

key_frames = key_frames[key_frames["snapFrame"].notna()].copy()

key_frames["endFrame"] = key_frames["passFrame"]
key_frames.loc[key_frames["endFrame"].isna(), "endFrame"] = key_frames["sackFrame"]
key_frames["endFrame"] = key_frames["endFrame"].fillna(key_frames["snapFrame"] + 30).astype(int)

tp = tp.merge(key_frames[["gameId", "playId", "snapFrame", "endFrame"]],
              on=["gameId", "playId"], how="inner")

tp_window = tp[(tp["frameId"] >= tp["snapFrame"]) & (tp["frameId"] <= tp["endFrame"])].copy()
print("tp_window shape:", tp_window.shape)


# =========================
# 5) IDENTIFY DEFENSIVE PASS RUSHERS AT SNAP
# =========================
snap = tp[tp["frameId"].eq(tp["snapFrame"])].copy()

ball_snap = snap[snap["teamAbbr"].eq("football")][["gameId", "playId", "x", "y"]].rename(
    columns={"x": "ball_x_snap", "y": "ball_y_snap"}
)
snap = snap.merge(ball_snap, on=["gameId", "playId"], how="inner")

def_snap = snap[
    (snap["teamAbbr"].eq(snap["defTeamAbbr"])) &
    (snap["nflId"].notna())
].copy()

if "position" not in def_snap.columns or "displayName" not in def_snap.columns:
    def_snap = def_snap.merge(players_small, on="nflId", how="left")

PASS_RUSH_POSITIONS = {"DE", "DT", "OLB", "EDGE"}
def_snap = def_snap[def_snap["position"].isin(PASS_RUSH_POSITIONS)].copy()

def_snap["dist_to_ball_x"] = (def_snap["x"] - def_snap["ball_x_snap"]).abs()
def_snap = def_snap[def_snap["dist_to_ball_x"] <= LOS_X_TOL].copy()

def_snap["rank_los"] = def_snap.groupby(["gameId", "playId"])["dist_to_ball_x"].rank(method="first")

rushers_snap = def_snap[
    def_snap["rank_los"] <= def_snap["numberOfPassRushers"]
][["gameId", "playId", "nflId", "displayName", "position", "numberOfPassRushers"]].drop_duplicates()

print("Rushers identified:", len(rushers_snap))
print("Rushers per play (describe):")
print(rushers_snap.groupby(["gameId", "playId"]).size().describe())


# =========================
# 5.1) RUSH-UNIT FEATURES AT SNAP (play_context)
# =========================
# ===== NEW: OFFENSIVE LINE LANDMARK ALIGNMENT =====
snap_all = tp[tp["frameId"].eq(tp["snapFrame"])].copy()

ball_snap_ctx = snap_all[snap_all["teamAbbr"].eq("football")][["gameId","playId","x","y"]].rename(
    columns={"x":"ball_x_snap","y":"ball_y_snap"}
)
snap_all = snap_all.merge(ball_snap_ctx, on=["gameId","playId"], how="left")

# Offensive players at snap
off_snap = snap_all[
    (snap_all["teamAbbr"].eq(snap_all["offTeamAbbr"])) &
    (snap_all["nflId"].notna())
].copy()

if "position" not in off_snap.columns:
    off_snap = off_snap.merge(
        players_small[["nflId","position"]],
        on="nflId",
        how="left"
    )

off_snap["los_dist"] = (off_snap["x"] - off_snap["ball_x_snap"]).abs()
off_los = off_snap[
    off_snap["position"].isin(["C","G","OG","T","OT","TE"]) &
    (off_snap["los_dist"] <= 1.5)
].copy()

# Defensive players at snap  <-- THIS WAS MISSING
def_snap_all = snap_all[
    (snap_all["teamAbbr"].eq(snap_all["defTeamAbbr"])) &
    (snap_all["nflId"].notna())
].copy()

if "position" not in def_snap_all.columns:
    def_snap_all = def_snap_all.merge(
        players_small[["nflId","position"]],
        on="nflId",
        how="left"
    )

def_snap_all["abs_y_from_ball"] = (def_snap_all["y"] - def_snap_all["ball_y_snap"]).abs()
def_snap_all["abs_x_from_ball"] = (def_snap_all["x"] - def_snap_all["ball_x_snap"]).abs()

rush_snap = def_snap_all.merge(
    rushers_snap[["gameId","playId","nflId"]],
    on=["gameId","playId","nflId"],
    how="inner"
).copy()

# ===== Alignment / location features at snap (per rusher) =====
# rush_snap currently has: gameId, playId, nflId, x, y, ball_x_snap, ball_y_snap

rush_snap["y_rel_snap"] = rush_snap["y"] - rush_snap["ball_y_snap"]
rush_snap["x_rel_snap"] = rush_snap["x"] - rush_snap["ball_x_snap"]

rush_snap["align_bucket"] = pd.cut(
    rush_snap["y_rel_snap"].abs(),
    bins=[-0.01, 1.5, 3.5, 10, 100],
    labels=["head_up", "inside_shade", "off_tackle", "wide9_plus"]
).astype("string")

rusher_alignment = rush_snap[
    ["gameId","playId","nflId","align_bucket","y_rel_snap","x_rel_snap"]
].drop_duplicates()

rush_snap["side"] = np.where(rush_snap["y"] >= rush_snap["ball_y_snap"], 1, -1)

play_context = (
    rush_snap.groupby(["gameId","playId"], as_index=False)
    .agg(
        num_rushers=("nflId","nunique"),
        front_width=("y", lambda s: float(np.max(s) - np.min(s))),
        mean_y_spread=("abs_y_from_ball","mean"),
        mean_x_los_dist=("abs_x_from_ball","mean"),
        left_right_balance=("side","mean"),
    )
)

# Ensure 1 row per play (some merges later depend on this)
play_context = play_context.drop_duplicates(["gameId","playId"]).copy()

play_context = play_context.merge(
    pass_plays_small[["gameId","playId","epa","passResult","numberOfPassRushers"]],
    on=["gameId","playId"],
    how="left"
)

print("play_context shape:", play_context.shape)
print(play_context.head())

# =========================
# 6) EXTRACT + NORMALIZE RUSHER TRAJECTORIES
# =========================
tp_rushers = tp_window.merge(
    rushers_snap[["gameId", "playId", "nflId"]],
    on=["gameId", "playId", "nflId"],
    how="inner"
)

tp_rushers = tp_rushers.merge(ball_snap, on=["gameId", "playId"], how="inner")

tp_rushers["x_rel"] = tp_rushers["x"] - tp_rushers["ball_x_snap"]
tp_rushers["y_rel"] = tp_rushers["y"] - tp_rushers["ball_y_snap"]


# =========================
# 7) RESAMPLE EACH RUSH TO FIXED LENGTH VECTOR
# =========================
def resample_to_k(df, k=K_FRAMES):
    df = df.sort_values("frameId")
    if len(df) < 2:
        return None
    t = np.linspace(0, 1, len(df))
    tq = np.linspace(0, 1, k)
    x = np.interp(tq, t, df["x_rel"].to_numpy())
    y = np.interp(tq, t, df["y_rel"].to_numpy())
    return np.concatenate([x, y])

X, meta_rows = [], []

for (gameId, playId, nflId), g in tp_rushers.groupby(["gameId", "playId", "nflId"]):
    vec = resample_to_k(g)
    if vec is None:
        continue
    X.append(vec)
    meta_rows.append((gameId, playId, nflId))

X = np.vstack(X) if X else np.empty((0, 2 * K_FRAMES))
meta = pd.DataFrame(meta_rows, columns=["gameId", "playId", "nflId"])
meta["orig_row"] = np.arange(len(meta))

# Merge outcomes + play_context + player identity
# play_context already has epa/passResult/numberOfPassRushers
meta = meta.merge(
    play_context[["gameId","playId","epa","passResult","numberOfPassRushers",
                  "num_rushers","front_width","mean_y_spread","mean_x_los_dist","left_right_balance"]],
    on=["gameId","playId"],
    how="left"
)
# Merge player identity
meta = meta.merge(players_small, on="nflId", how="left")

# Safety: guarantee epa exists
if "epa" not in meta.columns:
    meta["epa"] = np.nan

# compute def_effect
meta["def_effect"] = -meta["epa"]

# =========================
# FIX: EPA cleanup in meta (DO THIS ONCE)
# =========================
if "epa" not in meta.columns:
    meta["epa"] = np.nan

# If merges created epa_x/epa_y, combine them into meta["epa"]
if ("epa_x" in meta.columns) or ("epa_y" in meta.columns):
    # start with epa_x if present, otherwise NaN series
    meta["epa"] = meta.get("epa_x", pd.Series(index=meta.index, dtype="float64"))
    # fill missing with epa_y if present
    if "epa_y" in meta.columns:
        meta["epa"] = meta["epa"].combine_first(meta["epa_y"])

    meta.drop(columns=[c for c in ["epa_x","epa_y"] if c in meta.columns], inplace=True)

# Now compute defensive effect
meta["def_effect"] = -meta["epa"]

# Guarantee epa exists (robustness)
if "epa" not in meta.columns:
    meta["epa"] = np.nan

meta["def_effect"] = np.where(meta["epa"].notna(), -meta["epa"], np.nan)

print("Trajectory matrix X:", X.shape)
print(meta.head())

# --- EPA cleanup: ensure ONE epa column only ---
if "epa" not in meta.columns:
    # if merge created epa_x/epa_y, combine them
    if "epa_x" in meta.columns or "epa_y" in meta.columns:
        meta["epa"] = meta.get("epa_x")
        meta["epa"] = meta["epa"].combine_first(meta.get("epa_y"))
    else:
        meta["epa"] = np.nan

# drop any leftover suffix columns so I don't confuse MYSELF later
for c in ["epa_x","epa_y","passResult_x","passResult_y",
          "numberOfPassRushers_x","numberOfPassRushers_y"]:
    if c in meta.columns:
        meta.drop(columns=c, inplace=True)

# compute def_effect correctly
meta["def_effect"] = -meta["epa"]

# ===== OUTLIER REJECTION (trajectory-level) =====
# Compute max step size in the resampled series (rough jitter proxy)
X_xy = X.reshape(X.shape[0], 2, K_FRAMES)  # [n,2,K]
dx = np.diff(X_xy[:,0,:], axis=1)
dy = np.diff(X_xy[:,1,:], axis=1)
step = np.sqrt(dx**2 + dy**2)

meta["max_step"] = step.max(axis=1)
meta["mean_step"] = step.mean(axis=1)

# Drop extreme outliers (tune thresholds)
max_step_thresh = meta["max_step"].quantile(0.995)  # top 0.5%
keep_mask = meta["max_step"] <= max_step_thresh

X = X[keep_mask.to_numpy()]
meta = meta.loc[keep_mask].reset_index(drop=True)
meta["orig_row"] = np.arange(len(meta))
#look at label we give to code and make sure its align, make sure labels fit after velocity add.
# =========================
# 8) CLUSTER TRAJECTORIES
# =========================
if len(X) == 0:
    raise RuntimeError("No trajectories extracted. Try LOS_X_TOL=3.0 or reduce weeks to debug.")

# Directional clustering
Xz_dir = StandardScaler().fit_transform(X)
meta["cluster_dir"] = KMeans(
    n_clusters=CLUSTERS_K,
    random_state=42,
    n_init="auto"
).fit_predict(Xz_dir)

# Absolute-value clustering (recommended for left/right invariance)
Xz_abs = StandardScaler().fit_transform(np.abs(X))
meta["cluster_abs"] = KMeans(
    n_clusters=CLUSTERS_K,
    random_state=42,
    n_init="auto"
).fit_predict(Xz_abs)

cluster_names = {
    0: "Speed Rush (Wide Arc)",
    1: "Bull Rush (Power)",
    2: "Spin / Counter Rush",
    3: "Inside Pressure / Slant",
    4: "Explosive Burst",
    5: "Finesse Rush"
}

meta["cluster_abs"] = meta["cluster_abs"].astype(int)
meta["cluster_label"] = meta["cluster_abs"].map(cluster_names)

print("\nCluster counts (absolute-value clustering):")
print(meta["cluster_abs"].value_counts().sort_index())
 #(down and long distances, Velocity, are we literally spining from shoulders , effect outcome is situational stuff, post what is expected pressure, what would you have been expected to do, predictive analysis, postion on fields whats the pressure probablity, how confident , fixed legnth, fixed vector, controlling for off tempo plays scrambling, roll outs for standard drop back , is this a in general, alignment  Move toward more dynamic timewarping  dwt with clustering  startification and outlier rejection, reruning alanysis code with stratify  by player position and location on field , and)# PFF supplemental data and how the player was doubleteamed

print("cluster_velocity in meta:", "cluster_velocity" in meta.columns)
print("meta columns:", meta.columns.tolist())

# =========================
# 9) EFFECTIVENESS SUMMARY (EPA and -EPA)
# =========================
# def_effect already safe even if epa missing
print("\nMean -EPA by labeled cluster:")
print(meta.groupby("cluster_label")["def_effect"].mean().sort_values(ascending=False))


# =========================
# 10) PRESSURE + QB MOVEMENT (meta2)
# =========================
# A) Pick ONE QB per play (closest to ball at snap)
snap_rows = tp_window[tp_window["frameId"].eq(tp_window["snapFrame"])].copy()

ball_snap2 = snap_rows[snap_rows["teamAbbr"].eq("football")][["gameId","playId","x","y"]].rename(
    columns={"x":"ball_x_snap","y":"ball_y_snap"}
)

qb_rows = tp_window[
    (tp_window["teamAbbr"].eq(tp_window["offTeamAbbr"])) &
    (tp_window["position"].eq("QB")) &
    (tp_window["nflId"].notna())
].copy()

qb_snap = qb_rows.merge(
    snap_rows[["gameId","playId","nflId","x","y","snapFrame"]],
    on=["gameId","playId","nflId","snapFrame"],
    how="inner"
).merge(ball_snap2, on=["gameId","playId"], how="left")

qb_snap["dist_ball"] = np.sqrt(
    (qb_snap["x_x"]-qb_snap["ball_x_snap"])**2 + (qb_snap["y_x"]-qb_snap["ball_y_snap"])**2
)

qb_choice = (
    qb_snap.sort_values(["gameId","playId","dist_ball"])
          .drop_duplicates(["gameId","playId"])[["gameId","playId","nflId"]]
          .rename(columns={"nflId":"qbId"})
)

# Keep only chosen QB trajectory rows (all frames)
qb = qb_rows.merge(qb_choice, on=["gameId","playId"], how="inner")
qb = qb[qb["nflId"].eq(qb["qbId"])].copy()
qb = qb.sort_values(["gameId","playId","frameId"])

# B) QB movement metrics
qb["qb_dx"] = qb.groupby(["gameId","playId"])["x"].diff()
qb["qb_dy"] = qb.groupby(["gameId","playId"])["y"].diff()
qb["qb_step_dist"] = np.sqrt(qb["qb_dx"]**2 + qb["qb_dy"]**2)

qb_metrics = (
    qb.groupby(["gameId","playId"], as_index=False)
      .agg(
          qb_total_distance=("qb_step_dist","sum"),
          qb_lateral_sd=("y","std"),
          qb_depth_sd=("x","std"),
          qb_frames=("frameId","count")
      )
)
qb_metrics["qb_seconds_observed"] = qb_metrics["qb_frames"] / FPS

# Drop plays with missing QB tracking (no qb_metrics row)
valid_qb_plays = qb_metrics[["gameId","playId"]].drop_duplicates()
meta = meta.merge(valid_qb_plays, on=["gameId","playId"], how="inner")
X = X[meta["orig_row"].to_numpy()]
meta = meta.reset_index(drop=True)
meta["orig_row"] = np.arange(len(meta))

tp_rushers = tp_rushers.merge(valid_qb_plays, on=["gameId","playId"], how="inner")

print("qb_metrics shape:", qb_metrics.shape)
print(qb_metrics.head())

# C) Join rusher frames to QB frames, compute dist per frame
qb_window = qb[["gameId","playId","frameId","snapFrame","endFrame","x","y"]].rename(
    columns={"x":"qb_x","y":"qb_y"}
)

rush_frames = tp_rushers[["gameId","playId","nflId","frameId","snapFrame","endFrame","x","y"]].copy()

rush_qb = rush_frames.merge(
    qb_window,
    on=["gameId","playId","frameId","snapFrame","endFrame"],
    how="inner"
)

rush_qb["dist_to_qb"] = np.sqrt((rush_qb["x"]-rush_qb["qb_x"])**2 + (rush_qb["y"]-rush_qb["qb_y"])**2)
rush_qb = rush_qb[rush_qb["frameId"] <= (rush_qb["snapFrame"] + MAX_FRAMES)].copy()

# ===== Build DTW series per rusher-play using dist_to_qb (cheaper than 2D) =====
series = (
    rush_qb.sort_values(["gameId","playId","nflId","frameId"])
          .groupby(["gameId","playId","nflId"])["dist_to_qb"]
          .apply(lambda s: s.to_numpy())
)

dtw_meta = series.reset_index().rename(columns={"dist_to_qb":"dist_series"})



# D) time-to-pressure for two thresholds
def first_pressure(rush_qb_df, yards_thresh, label):
    within = rush_qb_df[rush_qb_df["dist_to_qb"] <= yards_thresh].copy()
    out = (
        within.groupby(["gameId","playId","nflId"], as_index=False)["frameId"]
              .min()
              .rename(columns={"frameId": f"pressFrame_{label}"})
    )
    out = out.merge(key_frames[["gameId","playId","snapFrame","endFrame"]], on=["gameId","playId"], how="left")
    out[f"time_to_pressure_{label}_sec"] = (out[f"pressFrame_{label}"] - out["snapFrame"]) / FPS
    out = out[out[f"pressFrame_{label}"] <= out["endFrame"]].copy()
    out[f"pressured_{label}"] = 1
    return out[["gameId","playId","nflId", f"time_to_pressure_{label}_sec", f"pressured_{label}"]]

press_imm = first_pressure(rush_qb, PRESSURE_YARDS_IMMEDIATE, "imm")
press_poc = first_pressure(rush_qb, PRESSURE_YARDS_POCKET, "pocket")

# E) Build meta2
meta2 = meta.merge(press_imm, on=["gameId","playId","nflId"], how="left")
meta2 = meta2.merge(press_poc, on=["gameId","playId","nflId"], how="left")
meta2 = meta2.merge(qb_metrics, on=["gameId","playId"], how="left")
meta2 = meta2.merge(rusher_alignment, on=["gameId","playId","nflId"], how="left")




meta2["pressured_imm"] = meta2["pressured_imm"].fillna(0).astype(int)
meta2["pressured_pocket"] = meta2["pressured_pocket"].fillna(0).astype(int)

early_summary = safe_groupby_agg(
    meta2,
    group_cols=["position", "align_bucket", "cluster_label"],
    agg_spec={
        "rushes": ("playId", "count"),
        "imm_pressure": ("pressured_imm", "mean"),
        "pocket_influence": ("pressured_pocket", "mean"),
        "strain": ("strain_mean", "mean"),
    }
)
display(early_summary)
# def_effect should now work (not NaN)
meta2["def_effect"] = -meta2["epa"]


# --- EPA cleanup in meta2 (DO THIS FIRST) ---
if "epa" not in meta2.columns:
    # if merge created epa_x/epa_y, combine them into epa
    if "epa_x" in meta2.columns or "epa_y" in meta2.columns:
        meta2["epa"] = meta2.get("epa_x")
        meta2["epa"] = meta2["epa"].combine_first(meta2.get("epa_y"))
    else:
        meta2["epa"] = np.nan

# if epa exists but is all NaN and epa_x/epa_y exist, repair it anyway
if meta2["epa"].isna().all() and ("epa_x" in meta2.columns or "epa_y" in meta2.columns):
    meta2["epa"] = meta2.get("epa_x")
    meta2["epa"] = meta2["epa"].combine_first(meta2.get("epa_y"))

# drop suffix columns so it stop tripping later
for c in ["epa_x", "epa_y"]:
    if c in meta2.columns:
        meta2.drop(columns=c, inplace=True)

# compute def_effect
meta2["def_effect"] = -meta2["epa"]


print("\nmeta2 created:", meta2.shape)
print("Immediate pressured counts:\n", meta2["pressured_imm"].value_counts())
print("Pocket pressured counts:\n", meta2["pressured_pocket"].value_counts())


model_df = meta2.copy()

features = [
    "position",
    "align_bucket",
    "num_rushers",
    "front_width",
    "mean_y_spread",
    "mean_x_los_dist",
    "qb_total_distance"
]

model_df = model_df.dropna(subset=["pressured_imm"])
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

X_model = model_df[features]
y_model = model_df["pressured_imm"]

cat_cols = ["position", "align_bucket"]
num_cols = [c for c in features if c not in cat_cols]

preprocess = ColumnTransformer([
    ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
    ("num", StandardScaler(), num_cols)
])

model = Pipeline([
    ("prep", preprocess),
    ("clf", LogisticRegression(max_iter=1000))
])

model.fit(X_model, y_model)

model_df["expected_pressure"] = model.predict_proba(X_model)[:, 1]
model_df["pressure_over_expected"] = (
    model_df["pressured_imm"] - model_df["expected_pressure"]

)

meta2 = meta2.drop(
    columns=["expected_pressure", "pressure_over_expected"],
    errors="ignore"
).merge(
    model_df[[
        "gameId", "playId", "nflId",
        "expected_pressure", "pressure_over_expected"
    ]],
    on=["gameId", "playId", "nflId"],
    how="left"
)

print(meta2[["expected_pressure", "pressure_over_expected"]].head())





# ============================================================
# ALIGNMENT VALIDATION PLOTS
# ============================================================

import matplotlib.pyplot as plt
import numpy as np

# Safety: ensure required columns exist
required_cols = ["align_bucket", "y_rel_snap"]
missing = [c for c in required_cols if c not in meta2.columns]
if missing:
    print("Missing required columns:", missing)
else:

    # ------------------------------------------------------------
    # 1) Distribution of alignment buckets
    # ------------------------------------------------------------
    print("\nAlignment bucket distribution:")
    print(meta2["align_bucket"].value_counts(normalize=True))

    # ------------------------------------------------------------
    # 2) Summary stats per bucket
    # ------------------------------------------------------------
    align_stats = (
        meta2.groupby("align_bucket")
             .agg(
                 count=("playId","count"),
                 mean_abs_y=("y_rel_snap", lambda s: np.abs(s).mean()),
                 p25=("y_rel_snap", lambda s: np.abs(s).quantile(0.25)),
                 p50=("y_rel_snap", lambda s: np.abs(s).quantile(0.50)),
                 p75=("y_rel_snap", lambda s: np.abs(s).quantile(0.75))
             )
             .sort_values("mean_abs_y")
    )

    print("\nAlignment bucket stats:")
    display(align_stats)

    # ------------------------------------------------------------
    # 3) Histogram overlay (separation check)
    # ------------------------------------------------------------
    sample = meta2.sample(n=min(5000, len(meta2)), random_state=42)

    plt.figure()
    for label, df in sample.groupby("align_bucket"):
        plt.hist(np.abs(df["y_rel_snap"]), bins=40, alpha=0.4, label=label)

    plt.legend()
    plt.title("Alignment Buckets vs Absolute Y Distance")
    plt.xlabel("Absolute Y distance from ball (yards)")
    plt.ylabel("Count")
    plt.show()

    # ------------------------------------------------------------
    # 4) Boxplot (clean separation view)
    # ------------------------------------------------------------
    plt.figure()
    meta2.boxplot(column="y_rel_snap", by="align_bucket")
    plt.title("Alignment Bucket vs Y Distance")
    plt.suptitle("")
    plt.ylabel("y_rel_snap")
    plt.xticks(rotation=30)
    plt.show()

    # ------------------------------------------------------------
    # 5) Alignment vs Position (football sanity check)
    # ------------------------------------------------------------
    print("\nAlignment vs Position:")
    display(
        pd.crosstab(
            meta2["position"],
            meta2["align_bucket"],
            normalize="index"
        )
    )

    # ------------------------------------------------------------
    # 6) Scatter plot (geometry check)
    # ------------------------------------------------------------
    sample2 = meta2.sample(n=min(3000, len(meta2)), random_state=42)

    plt.figure()
    for label, df in sample2.groupby("align_bucket"):
        plt.scatter(
            df["y_rel_snap"],
            df.get("x_rel_snap", 0),  # safe fallback
            alpha=0.4,
            label=label
        )

    plt.axvline(0, linestyle="--")
    plt.title("Rusher Alignment Buckets (Y vs X)")
    plt.xlabel("Horizontal distance (y_rel)")
    plt.ylabel("Depth from LOS (x_rel)")
    plt.legend()
    plt.show()

    # ------------------------------------------------------------
    # 7) Mean alignment per bucket (quick numeric check)
    # ------------------------------------------------------------
    print("\nMean absolute alignment by bucket:")
    display(
        meta2.groupby("align_bucket")["y_rel_snap"]
             .apply(lambda s: np.abs(s).mean())
             .sort_values()
    )


# ============================================================
# STRAIN  — frame-level + rollups
# Requires: rush_qb with columns [gameId, playId, nflId, frameId, dist_to_qb]
# Also requires: meta2 with [gameId, playId, nflId, pressured_imm, ...]
# ============================================================

DT = 1.0 / FPS  # FPS = 10.0 in BDB tracking

# 1) Frame-level STRAIN (a "closing speed" normalized by current distance)
rush_qb = rush_qb.sort_values(["gameId", "playId", "nflId", "frameId"]).copy()
rush_qb["dist_prev"] = rush_qb.groupby(["gameId","playId","nflId"])["dist_to_qb"].shift(1)
rush_qb["d_dist"] = rush_qb["dist_to_qb"] - rush_qb["dist_prev"]

# STRAIN_t = -(d_dist/dt) / dist_to_qb
rush_qb["strain"] = -(rush_qb["d_dist"] / DT) / rush_qb["dist_to_qb"]

# Clean up
rush_qb.loc[rush_qb["dist_to_qb"] <= 0, "strain"] = np.nan
rush_qb.loc[rush_qb["dist_prev"].isna(), "strain"] = np.nan

# Optional: clip extreme noise spikes (tracking jitter can blow this up)
rush_qb["strain"] = rush_qb["strain"].clip(lower=-20, upper=20)

# 2) Roll up STRAIN per (play, rusher)
strain_rusher_play = (
    rush_qb.groupby(["gameId","playId","nflId"], as_index=False)
           .agg(
               strain_mean=("strain","mean"),
               strain_p90=("strain", lambda s: s.quantile(0.90)),
               strain_peak=("strain","max"),
               strain_frames=("strain","count")
           )
)

# 3) Merge into meta2
meta2 = meta2.merge(strain_rusher_play, on=["gameId","playId","nflId"], how="left")


meta2.groupby(["position","align_bucket","cluster_label"]).agg(
    rushes=("playId","count"),
    imm_pressure=("pressured_imm","mean"),
    pocket_influence=("pressured_pocket","mean"),
    strain=("strain_mean","mean")
).reset_index()

needed_cols = ["gameId", "playId", "nflId", "cluster_velocity", "cluster_label_velocity"]
missing_in_meta = [c for c in needed_cols if c not in meta.columns]

if missing_in_meta:
    print("Missing in meta:", missing_in_meta)
    print("Falling back to cluster_abs / cluster_label")
    meta["cluster_velocity"] = meta.get("cluster_velocity", meta["cluster_abs"])
    meta["cluster_label_velocity"] = meta.get("cluster_label_velocity", meta["cluster_label"])

meta2 = meta2.drop(
    columns=["cluster_velocity", "cluster_label_velocity"],
    errors="ignore"
).merge(
    meta[["gameId", "playId", "nflId", "cluster_velocity", "cluster_label_velocity"]],
    on=["gameId", "playId", "nflId"],
    how="left"
)

print("cluster_label_velocity in meta2:", "cluster_label_velocity" in meta2.columns)
print(meta2[["cluster_velocity", "cluster_label_velocity"]].head())

cluster_eval = (
    meta2.groupby("cluster_label_velocity")
         .agg(
             rushes=("playId","count"),
             actual_pressure=("pressured_imm","mean"),
             expected_pressure=("expected_pressure","mean"),
             over_expected=("pressure_over_expected","mean"),
             strain=("strain_mean","mean")
         )
         .sort_values("over_expected", ascending=False)
)

display(cluster_eval)

eval_spec = {
    "rushes": ("playId","count"),
    "actual_pressure": ("pressured_imm","mean"),
    "expected_pressure": ("expected_pressure","mean"),
    "over_expected": ("pressure_over_expected","mean"),
}

if "strain_mean" in meta2.columns:
    eval_spec["strain"] = ("strain_mean","mean")

cluster_eval = (
    safe_groupby_agg(
        meta2,
        group_cols=["cluster_label_velocity"],
        agg_spec=eval_spec
    )
    .sort_values("over_expected", ascending=False)
)

display(cluster_eval)

# Make sure velocity labels are present in meta2
needed_cols = ["gameId", "playId", "nflId", "cluster_velocity", "cluster_label_velocity"]
missing_in_meta = [c for c in needed_cols if c not in meta.columns]

if missing_in_meta:
    print("Missing in meta:", missing_in_meta)
    print("Falling back to cluster_abs / cluster_label")
    meta["cluster_velocity"] = meta.get("cluster_velocity", meta["cluster_abs"])
    meta["cluster_label_velocity"] = meta.get("cluster_label_velocity", meta["cluster_label"])

meta2 = meta2.drop(
    columns=["cluster_velocity", "cluster_label_velocity"],
    errors="ignore"
).merge(
    meta[["gameId", "playId", "nflId", "cluster_velocity", "cluster_label_velocity"]],
    on=["gameId", "playId", "nflId"],
    how="left"
)

# Safe cluster evaluation
eval_spec = {
    "rushes": ("playId", "count"),
    "actual_pressure": ("pressured_imm", "mean"),
    "expected_pressure": ("expected_pressure", "mean"),
    "over_expected": ("pressure_over_expected", "mean"),
}

if "strain_mean" in meta2.columns:
    eval_spec["strain"] = ("strain_mean", "mean")

cluster_eval = (
    safe_groupby_agg(
        meta2,
        group_cols=["cluster_label_velocity"],
        agg_spec=eval_spec
    )
)

if "over_expected" in cluster_eval.columns:
    cluster_eval = cluster_eval.sort_values("over_expected", ascending=False)

display(cluster_eval)
# ============================================================
# 10B) MULTIVARIATE DTW + STATIC CONTEXT DISTANCE
# Paste this AFTER:
#   meta2 = meta2.merge(strain_rusher_play, on=["gameId","playId","nflId"], how="left")
# and BEFORE your final DTW summaries / plots
# ============================================================

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import pairwise_distances
from sklearn.cluster import AgglomerativeClustering

# ------------------------------------------------------------
# A) Build alignment / location features at snap (rusher-level)
# ------------------------------------------------------------
# Uses rush_snap from your earlier snap-based section if it still exists.
# If not, rebuild it from tp + rushers_snap.

if "rush_snap" not in globals():
    snap_all = tp[tp["frameId"].eq(tp["snapFrame"])].copy()

    ball_snap_ctx = snap_all[snap_all["teamAbbr"].eq("football")][["gameId","playId","x","y"]].rename(
        columns={"x":"ball_x_snap","y":"ball_y_snap"}
    )
    snap_all = snap_all.merge(ball_snap_ctx, on=["gameId","playId"], how="left")

    def_snap_all = snap_all[
        (snap_all["teamAbbr"].eq(snap_all["defTeamAbbr"])) &
        (snap_all["nflId"].notna())
    ].copy()

    # ✅ FIX: keep this INSIDE the block
    if "position" not in def_snap_all.columns:
        def_snap_all = def_snap_all.merge(
            players_small[["nflId","position"]],
            on="nflId",
            how="left"
        )

    rush_snap = def_snap_all.merge(
        rushers_snap[["gameId","playId","nflId"]],
        on=["gameId","playId","nflId"],
        how="inner"
    ).copy()

# relative coordinates
rush_snap["y_rel_snap"] = rush_snap["y"] - rush_snap["ball_y_snap"]
rush_snap["x_rel_snap"] = rush_snap["x"] - rush_snap["ball_x_snap"]

# keep only rushers near LOS
rush_snap["los_dist"] = rush_snap["x_rel_snap"].abs()
rush_snap = rush_snap[rush_snap["los_dist"] <= 1.5].copy()

# alignment buckets
rush_snap["align_bucket"] = pd.cut(
    rush_snap["y_rel_snap"].abs(),
    bins=[-0.01, 1.5, 3.5, 6.0, 100],
    labels=["head_up", "inside_shade", "off_tackle", "wide9_plus"]
).astype("string")

rusher_alignment = rush_snap[
    ["gameId","playId","nflId","align_bucket","y_rel_snap","x_rel_snap","los_dist"]
].drop_duplicates()

# Merge alignment into meta2 if not already there
align_cols_to_add = [c for c in ["align_bucket","y_rel_snap","x_rel_snap"] if c not in meta2.columns]
if align_cols_to_add:
    meta2 = meta2.merge(
        rusher_alignment[["gameId","playId","nflId"] + align_cols_to_add],
        on=["gameId","playId","nflId"],
        how="left"
    )

# ------------------------------------------------------------
# B) Add richer frame-level movement features to rush_qb
# ------------------------------------------------------------
# We need ball-relative x/y, velocity, acceleration, heading, change of heading.

ball_lookup = ball_snap.rename(columns={"ball_x_snap":"bx","ball_y_snap":"by"})
rush_qb = rush_qb.merge(ball_lookup, on=["gameId","playId"], how="left")

rush_qb["x_rel"] = rush_qb["x"] - rush_qb["bx"]
rush_qb["y_rel"] = rush_qb["y"] - rush_qb["by"]

DT_FRAME = 1.0 / FPS

rush_qb = rush_qb.sort_values(["gameId","playId","nflId","frameId"]).copy()
rush_qb["dx"] = rush_qb.groupby(["gameId","playId","nflId"])["x"].diff()
rush_qb["dy"] = rush_qb.groupby(["gameId","playId","nflId"])["y"].diff()

rush_qb["speed_xy"] = np.sqrt(rush_qb["dx"]**2 + rush_qb["dy"]**2) / DT_FRAME
rush_qb["accel_xy"] = rush_qb.groupby(["gameId","playId","nflId"])["speed_xy"].diff() / DT_FRAME

rush_qb["heading"] = np.arctan2(rush_qb["dy"], rush_qb["dx"])
rush_qb["d_heading"] = rush_qb.groupby(["gameId","playId","nflId"])["heading"].diff()

# wrap angle difference to [-pi, pi]
rush_qb["d_heading"] = ((rush_qb["d_heading"] + np.pi) % (2 * np.pi)) - np.pi

# optional additional curvature proxy
rush_qb["abs_d_heading"] = rush_qb["d_heading"].abs()

# ===== Velocity summary features per rusher-play =====
vel_features = (
    rush_qb.groupby(["gameId","playId","nflId"], as_index=False)
           .agg(
               max_speed=("speed_xy","max"),
               mean_speed=("speed_xy","mean"),
               max_accel=("accel_xy","max"),
               mean_accel=("accel_xy","mean"),
               mean_curvature=("abs_d_heading","mean")
           )
)
meta = meta.merge(
    vel_features,
    on=["gameId","playId","nflId"],
    how="left"
)
meta2 = meta2.merge(
    vel_features,
    on=["gameId","playId","nflId"],
    how="left"
)
# ===== NEW: VELOCITY-AWARE CLUSTERING =====
# Ensure base trajectory matrix exists
if "X" not in globals():
    raise ValueError("X (trajectory matrix) not found — clustering cannot proceed.")

X_base = np.abs(X)

vel_cols = ["max_speed","mean_speed","max_accel","mean_accel","mean_curvature"]

missing_vel = [c for c in vel_cols if c not in meta.columns]
print("Missing velocity columns in meta:", missing_vel)
print("Current meta columns:", meta.columns.tolist())

if missing_vel:
    raise ValueError(
        "Velocity-aware clustering is running before velocity features were created and merged into meta."
    )

vel_matrix = meta[vel_cols].fillna(0).to_numpy()

X_scaled = StandardScaler().fit_transform(X_base)
vel_scaled = StandardScaler().fit_transform(vel_matrix)

gamma = 0.3
X_combined = np.hstack([X_scaled, gamma * vel_scaled])

meta["cluster_velocity"] = KMeans(
    n_clusters=CLUSTERS_K,
    random_state=42,
    n_init="auto"
).fit_predict(X_combined)

cluster_profile = (
    meta.groupby("cluster_velocity")
        .agg(
            avg_speed=("mean_speed","mean"),
            avg_curvature=("mean_curvature","mean"),
            avg_alignment=("orig_row", lambda idx: np.abs(X[idx.to_numpy(), K_FRAMES:]).mean())
        )
)
meta2 = meta2.drop(
    columns=["cluster_velocity", "cluster_label_velocity"],
    errors="ignore"
).merge(
    meta[["gameId","playId","nflId","cluster_velocity","cluster_label_velocity"]],
    on=["gameId","playId","nflId"],
    how="left"
)
# =========================
# VELOCITY CLUSTER LABELING (CORRECT LOCATION)
# =========================

cluster_profile = (
    meta.groupby("cluster_velocity")
        .agg(
            avg_speed=("mean_speed","mean"),
            avg_curvature=("mean_curvature","mean"),
            avg_alignment=("orig_row", lambda idx: np.abs(X[idx.to_numpy(), K_FRAMES:]).mean())
        )
)

def assign_cluster_label(row):
    speed = row["avg_speed"]
    curve = row["avg_curvature"]
    align = row["avg_alignment"]

    if speed > 6 and align > 3:
        return "Explosive Speed Rush"
    elif curve > 1.0:
        return "Spin / Counter Rush"
    elif align < 2:
        return "Interior Rush"
    elif speed < 5:
        return "Power Bull Rush"
    else:
        return "Controlled Edge Rush"

cluster_profile["cluster_label_velocity"] = cluster_profile.apply(assign_cluster_label, axis=1)

cluster_label_map = cluster_profile["cluster_label_velocity"].to_dict()

meta["cluster_label_velocity"] = meta["cluster_velocity"].map(cluster_label_map)

print("\nVelocity cluster labels:")
display(cluster_profile)

meta2 = meta2.drop(
    columns=["cluster_velocity", "cluster_label_velocity"],
    errors="ignore"
).merge(
    meta[["gameId","playId","nflId","cluster_velocity","cluster_label_velocity"]],
    on=["gameId","playId","nflId"],
    how="left"
)
# ------------------------------------------------------------
# C) Frame-level DTW feature set
# ------------------------------------------------------------
dtw_feature_cols = [
    "x_rel",
    "y_rel",
    "dist_to_qb",
    "dx",
    "dy",
    "speed_xy",
    "accel_xy",
    "heading",
    "abs_d_heading"
]

# clean infinities
rush_qb[dtw_feature_cols] = rush_qb[dtw_feature_cols].replace([np.inf, -np.inf], np.nan)

# Standardize numeric features before DTW
frame_scaler = StandardScaler()

valid_frame_rows = rush_qb[dtw_feature_cols].dropna()
if len(valid_frame_rows) == 0:
    raise ValueError("No valid rows found for DTW feature scaling.")

frame_scaler.fit(valid_frame_rows)

rush_qb_scaled = rush_qb.copy()
rush_qb_scaled[dtw_feature_cols] = frame_scaler.transform(
    rush_qb_scaled[dtw_feature_cols].fillna(0)
)

# ------------------------------------------------------------
# D) Build multivariate time-series per rusher-play
# Each sequence is [T frames, F features]
# ------------------------------------------------------------
multi_dtw_meta = (
    rush_qb_scaled
    .sort_values(["gameId","playId","nflId","frameId"])
    .groupby(["gameId","playId","nflId"])
    .apply(lambda g: g[dtw_feature_cols].to_numpy())
    .reset_index(name="feature_seq")
)

# ------------------------------------------------------------
# E) Multivariate DTW distance function
# ------------------------------------------------------------
def multivariate_dtw(seq1, seq2):
    """
    seq1: np.array [T1, F]
    seq2: np.array [T2, F]
    returns scalar DTW distance
    """
    n, m = len(seq1), len(seq2)
    Dmat = np.full((n + 1, m + 1), np.inf)
    Dmat[0, 0] = 0.0

    for i in range(1, n + 1):
        for j in range(1, m + 1):
            cost = np.linalg.norm(seq1[i - 1] - seq2[j - 1])
            Dmat[i, j] = cost + min(
                Dmat[i - 1, j],
                Dmat[i, j - 1],
                Dmat[i - 1, j - 1]
            )

    return Dmat[n, m]

# ------------------------------------------------------------
# F) Build subset for DTW clustering
# Change POSITION_SUBSET to "OLB", "DT", etc. as needed
# ------------------------------------------------------------
POSITION_SUBSET = "DE"
DTW_SAMPLE_N = 250
DTW_K = 6

sub = meta2[
    meta2["position"].eq(POSITION_SUBSET)
].dropna(subset=["strain_mean"]).copy()

if len(sub) == 0:
    raise ValueError(f"No rows available for DTW subset position={POSITION_SUBSET}")

sub = sub.sample(n=min(DTW_SAMPLE_N, len(sub)), random_state=42).copy()

sub_key = list(zip(sub["gameId"], sub["playId"], sub["nflId"]))

seq_lookup = multi_dtw_meta.set_index(["gameId","playId","nflId"])["feature_seq"]

available_keys = [k for k in sub_key if k in seq_lookup.index]
sub = sub.set_index(["gameId","playId","nflId"]).loc[available_keys].reset_index()

sub_sequences = [seq_lookup.loc[k] for k in available_keys]

# ------------------------------------------------------------
# G) DTW distance matrix on multivariate sequences
# ------------------------------------------------------------
n = len(sub_sequences)
D_dtw = np.zeros((n, n), dtype=float)

for i in range(n):
    for j in range(i + 1, n):
        D_dtw[i, j] = multivariate_dtw(sub_sequences[i], sub_sequences[j])
        D_dtw[j, i] = D_dtw[i, j]

# ------------------------------------------------------------
# H) Static context matrix (position/alignment/context)
# Important: strings go HERE, not inside the DTW sequence itself
# ------------------------------------------------------------
static_cols = []
for c in ["position", "align_bucket", "down_dist_bucket", "pass_obvious"]:
    if c in sub.columns:
        static_cols.append(c)

if len(static_cols) > 0:
    static_encoded = pd.get_dummies(sub[static_cols], dummy_na=True)
    D_static = pairwise_distances(static_encoded, metric="euclidean")
else:
    D_static = np.zeros_like(D_dtw)

# ------------------------------------------------------------
# I) Combine DTW movement distance + static context distance
# Tune weights as needed
# ------------------------------------------------------------
alpha = 0.85   # movement sequence weight
beta  = 0.15   # static context weight

D_combined = alpha * D_dtw + beta * D_static

# ------------------------------------------------------------
# J) Cluster the combined distance matrix
# ------------------------------------------------------------
dtw_clusters = AgglomerativeClustering(
    n_clusters=DTW_K,
    metric="precomputed",
    linkage="average"
).fit_predict(D_combined)

sub["cluster_dtw_multi"] = dtw_clusters

# Merge back into meta2
meta2 = meta2.merge(
    sub[["gameId","playId","nflId","cluster_dtw_multi"]],
    on=["gameId","playId","nflId"],
    how="left"
)

print("DTW multivariate subset shape:", sub.shape)
print("DTW multivariate cluster counts:")
print(sub["cluster_dtw_multi"].value_counts().sort_index())

# ------------------------------------------------------------
# K) Summarize DTW multivariate clusters
# ------------------------------------------------------------
dtw_multi_summary = (
    sub.groupby("cluster_dtw_multi", as_index=False)
       .agg(
           rushes=("playId","count"),
           imm_pressure=("pressured_imm","mean"),
           pocket_influence=("pressured_pocket","mean"),
           avg_ttp_imm=("time_to_pressure_imm_sec","mean"),
           avg_ttp_pocket=("time_to_pressure_pocket_sec","mean"),
           mean_def_effect=("def_effect","mean"),
           mean_strain=("strain_mean","mean")
       )
       .sort_values("mean_def_effect", ascending=False)
)

print("\nDTW Multivariate Cluster Summary")
display(dtw_multi_summary)

if all(c in meta2.columns for c in ["cluster_label_velocity", "double_teamed", "pressure_over_expected"]):
    display(
        meta2.groupby("cluster_label_velocity")[["double_teamed", "pressure_over_expected"]].mean()
    )
else:
    print("Skipping double-team summary; required columns not present.")
# ------------------------------------------------------------
# L) Optional: compare original KMeans cluster vs DTW multivariate cluster
# ------------------------------------------------------------
if "cluster_abs" in sub.columns:
    overlap = pd.crosstab(sub["cluster_abs"], sub["cluster_dtw_multi"])
    print("\nKMeans vs DTW multivariate cluster overlap")
    display(overlap)

# ===== DTW distance matrix (SUBSET) =====
# Example: only DEs and sample 600 rushes
sub = meta2[meta2["position"].eq("DE")].dropna(subset=["strain_mean"]).copy()
sub = sub.sample(n=min(600, len(sub)), random_state=42)

sub_key = list(zip(sub["gameId"], sub["playId"], sub["nflId"]))
sub_series = dtw_meta.set_index(["gameId","playId","nflId"]).loc[sub_key]["dist_series"].tolist()

# pairwise DTW distances
n = len(sub_series)
D = np.zeros((n,n), dtype=float)
for i in range(n):
    for j in range(i+1, n):
        D[i,j] = dtw.distance_fast(sub_series[i], sub_series[j])
        D[j,i] = D[i,j]

from sklearn.cluster import AgglomerativeClustering

dtw_clusters = AgglomerativeClustering(
    n_clusters=6,
    metric="precomputed",
    linkage="average"
).fit_predict(D)

# Label DTW clusters
if "cluster_dtw" not in sub.columns:
    sub["cluster_dtw"] = dtw_clusters
dtw_cluster_names = {
    0: "DTW Cluster 0",
    1: "DTW Cluster 1",
    2: "DTW Cluster 2",
    3: "DTW Cluster 3",
    4: "DTW Cluster 4",
    5: "DTW Cluster 5",
}
print(sub.columns)
sub["cluster_dtw_label"] = sub["cluster_dtw"].map(dtw_cluster_names)

# Merge DTW labels into meta2
meta2 = meta2.drop(columns=["cluster_dtw", "cluster_dtw_label"], errors="ignore").merge(
    sub[["gameId", "playId", "nflId", "cluster_dtw", "cluster_dtw_label"]],
    on=["gameId", "playId", "nflId"],
    how="left"
)

# DTW summary table
dtw_summary = (
    meta2.dropna(subset=["cluster_dtw"])
         .groupby(["cluster_dtw", "cluster_dtw_label"], as_index=False)
         .agg(
             rushes=("playId", "count"),
             pressure_rate_imm=("pressured_imm", "mean"),
             avg_ttp_imm=("time_to_pressure_imm_sec", "mean"),
             pressure_rate_pocket=("pressured_pocket", "mean"),
             avg_ttp_pocket=("time_to_pressure_pocket_sec", "mean"),
             strain_mean=("strain_mean", "mean"),
             mean_epa=("epa", "mean"),
             mean_def_effect=("def_effect", "mean"),
             qb_move=("qb_total_distance", "mean"),
         )
         .sort_values("mean_def_effect", ascending=False)
)

display(dtw_summary)
plt.figure()
plt.bar(dtw_summary["cluster_dtw_label"], dtw_summary["pressure_rate_imm"])
plt.title("Immediate Pressure Rate by DTW Cluster")
plt.xlabel("DTW Cluster")
plt.ylabel("Immediate pressure rate")
plt.xticks(rotation=30, ha="right")
plt.show()

plt.figure()
plt.bar(dtw_summary["cluster_dtw_label"], dtw_summary["pressure_rate_pocket"])
plt.title("Pocket Influence Rate by DTW Cluster")
plt.xlabel("DTW Cluster")
plt.ylabel("Pocket influence rate")
plt.xticks(rotation=30, ha="right")
plt.show()

meta2 = meta2.drop(
    columns=["strain_mean", "strain_p90", "strain_peak", "strain_frames"],
    errors="ignore"
).merge(
    strain_rusher_play,
    on=["gameId","playId","nflId"],
    how="left"
)
def run_kmeans_for_subset(df_subset, X_subset, k=6):
    Xz = StandardScaler().fit_transform(np.abs(X_subset))
    labels = KMeans(n_clusters=k, random_state=42, n_init="auto").fit_predict(Xz)
    return labels

# Example: cluster separately by position
meta2["cluster_abs_by_pos"] = -1
for pos in ["DE","OLB","DT","NT"]:
    idx = meta2["position"].eq(pos).to_numpy()
    if idx.sum() < 200:
        continue
    meta2.loc[idx, "cluster_abs_by_pos"] = run_kmeans_for_subset(meta2.loc[idx], X[idx], k=6)

# 4) “Best STRAIN” by position (DE/OLB/DT/NT)
# NOTE:  data may not have NT for all nose tackles; sometimes it's DT.
KEEP_POS = {"DE","OLB","DT","NT"}

pos_leaderboard = (
    meta2[meta2["position"].isin(KEEP_POS)]
      .groupby(["position","displayName","nflId"], as_index=False)
      .agg(
          rushes=("playId","count"),
          strain_mean=("strain_mean","mean"),
          strain_p90=("strain_p90","mean"),
          strain_peak=("strain_peak","mean"),
          imm_pressure_rate=("pressured_imm","mean"),
          pocket_influence_rate=("pressured_pocket","mean"),
          avg_ttp_imm=("time_to_pressure_imm_sec","mean"),
      )
)

dtw_vs_kmeans = (
    meta2.dropna(subset=["cluster_dtw"])
         .groupby(["cluster_label", "cluster_dtw_label"], as_index=False)
         .agg(rushes=("playId", "count"))
         .sort_values("rushes", ascending=False)
)

display(dtw_vs_kmeans.head(20))
# Filter out tiny samples so “best” is meaningful
MIN_RUSHES = 30
pos_leaderboard = pos_leaderboard[pos_leaderboard["rushes"] >= MIN_RUSHES].copy()

# Top 10 by STRAIN per position
top_strain_by_pos = (
    pos_leaderboard.sort_values(["position","strain_mean"], ascending=[True, False])
                  .groupby("position", as_index=False)
                  .head(10)
)

display(top_strain_by_pos)

# 5) Compare STRAIN vs current pipeline metrics (quick sanity)
compare_cols = ["strain_mean","strain_p90","imm_pressure_rate","pocket_influence_rate","qb_total_distance","epa"]
cols_present = [c for c in compare_cols if c in meta2.columns]
display(meta2[cols_present].corr(numeric_only=True))
# =========================
# 11) PRESSURE SUMMARY BY CLUSTER + PLOTS (Immediate primary)
# =========================
cluster_pressure = (
    meta2.groupby("cluster_label")
         .agg(
             rushes=("nflId", "count"),
             pressure_rate_imm=("pressured_imm", "mean"),
             avg_ttp_imm=("time_to_pressure_imm_sec", "mean"),
             median_ttp_imm=("time_to_pressure_imm_sec", "median"),
             pressure_rate_pocket=("pressured_pocket", "mean"),
             avg_ttp_pocket=("time_to_pressure_pocket_sec", "mean"),
             median_ttp_pocket=("time_to_pressure_pocket_sec", "median"),
         )
)
label_order = list(cluster_names.values())
plot_df = cluster_pressure.reindex(label_order)

plt.figure()
plt.bar(plot_df.index, plot_df["pressure_rate_imm"].values)
plt.title("Immediate Pressure Rate (<= 2.0 yds) by Rush Type")
plt.xlabel("Rush Type")
plt.ylabel("Immediate pressure rate")
plt.xticks(rotation=30, ha="right")
plt.show()

plt.figure()
plt.bar(plot_df.index, plot_df["pressure_rate_pocket"].values)
plt.title("Pocket Influence Rate (<= 3.5 yds) by Rush Type")
plt.xlabel("Rush Type")
plt.ylabel("Pocket influence rate")
plt.xticks(rotation=30, ha="right")
plt.show()

print(plot_df)


# =========================
# 12) PLAY-LEVEL DATASET (play_level) + play_pressure flags
# =========================
play_pressure = (
    meta2.groupby(["gameId","playId"], as_index=False)
        .agg(
            any_pressure_imm=("pressured_imm", "max"),
            any_pressure_pocket=("pressured_pocket", "max"),
            avg_ttp_imm=("time_to_pressure_imm_sec", "mean"),
            avg_ttp_pocket=("time_to_pressure_pocket_sec", "mean"),
        )
)

# Build play_level from play_context (owns front_width/num_rushers) + qb_metrics + play_pressure
play_level = play_context.merge(qb_metrics, on=["gameId","playId"], how="left")
play_level = play_level.merge(play_pressure, on=["gameId","playId"], how="left")

print("play_level shape:", play_level.shape)
print(play_level.head())


# =========================
# 13) PLAYER RUSH-TYPE FINGERPRINTS (scouting profiles)
# Immediate primary; pocket secondary
# =========================
# Ensure label exists
if "cluster_label" not in meta2.columns:
    meta2["cluster_label"] = meta2["cluster_abs"].map(cluster_names)

# Rush-type distribution proportions
dist = (
    meta2.pivot_table(index=["displayName","nflId"],
                      columns="cluster_label",
                      values="playId",
                      aggfunc="count",
                      fill_value=0)
)
dist = dist.div(dist.sum(axis=1), axis=0).reset_index()

player_fingerprints = (
    meta2.groupby(["displayName","nflId"], as_index=False)
         .agg(
             rushes=("playId","count"),
             pressure_rate_imm=("pressured_imm","mean"),
             avg_ttp_imm=("time_to_pressure_imm_sec","mean"),
             pressure_rate_pocket=("pressured_pocket","mean"),
             avg_ttp_pocket=("time_to_pressure_pocket_sec","mean"),
             mean_def_effect=("def_effect","mean"),
         )
)

player_fingerprints = player_fingerprints.merge(dist, on=["displayName","nflId"], how="left")
player_fingerprints = player_fingerprints.sort_values(["pressure_rate_imm","rushes"], ascending=[False, False])

print(player_fingerprints.head(10))


# ============================================================
# DROP-IN PATCH: add these 3 replacements into the big code
# (1) Replace safe_groupby_agg
# (2) Convert front_width_bucket to string before interaction
# (3) Replace interaction aggregation block
# ============================================================

# ---------- (1) REPLACE  safe_groupby_agg with this ----------
def safe_groupby_agg(df, group_cols, agg_spec, sort=False, observed=True):
    """
    Returns a normal DataFrame with group_cols as columns.
    Avoids pandas as_index=False insertion bug by using reset_index().
    Any agg whose source column is missing is dropped automatically.
    """
    keep = {}
    missing = []
    for out_name, (src_col, func) in agg_spec.items():
        if src_col in df.columns:
            keep[out_name] = (src_col, func)
        else:
            missing.append((out_name, src_col))
    if missing:
        print("⚠️ Dropping missing agg columns:", missing)

    g = df.groupby(group_cols, as_index=True, sort=sort, observed=observed).agg(**keep)
    return g.reset_index()


# ---------- ... keep  code above ... ----------


# ============================================================
# STEP 14) INTERACTION TEST: Rush type effectiveness by front context
# ============================================================

# --- Create rusher_level DataFrame for interaction analysis ---
# This DataFrame combines rusher-specific data (from meta2) with play-level context (from play_level)
# It's needed for bucketing 'front_width' and then grouping for interaction analysis.
rusher_level = meta2.merge(play_level, on=["gameId", "playId"], how="left", suffixes=('', '_play'))

# --- Bucket front width robustly  ---
fw = rusher_level["front_width"]
try:
    rusher_level["front_width_bucket"] = pd.qcut(
        fw, 3, labels=["tight","medium","wide"], duplicates="drop"
    )
except ValueError:
    fw_nonnull = fw.dropna()
    if fw_nonnull.nunique() < 3:
        rusher_level["front_width_bucket"] = np.where(fw.notna(), "medium", np.nan)
    else:
        rusher_level["front_width_bucket"] = pd.cut(fw, bins=3, labels=["tight","medium","wide"])

# ---------- (2) ADD THIS right before computing interaction ----------
# qcut returns a Categorical; converting to string prevents the
# pandas groupby(as_index=False) insertion length bug.
if "front_width_bucket" in rusher_level.columns:
    rusher_level["front_width_bucket"] = rusher_level["front_width_bucket"].astype("string")

# ---------- (3) REPLACE the interaction = (...) block with this ----------
interaction_spec = {
    "rushes": ("nflId", "count"),
    "pressure_rate_imm": ("pressured_imm", "mean"),
    "pressure_rate_pocket": ("pressured_pocket", "mean"),
    "avg_def_effect": ("def_effect", "mean"),
    "avg_qb_move": ("qb_total_distance", "mean"),
}

interaction = (
    safe_groupby_agg(
        rusher_level,
        group_cols=["cluster_label", "front_width_bucket"],
        agg_spec=interaction_spec,
        observed=True
    )
    .sort_values(["cluster_label", "front_width_bucket"])
)

print("interaction shape:", interaction.shape)
display(interaction.head(20))


# =========================
# 15) CLUSTER SUMMARY: effective vs ineffective (Immediate primary)
# =========================
cluster_summary = (
    meta2.groupby("cluster_label", as_index=False)
         .agg(
             rushes=("nflId", "count"),
             pressure_rate_imm=("pressured_imm", "mean"),
             avg_ttp_imm=("time_to_pressure_imm_sec", "mean"),
             pressure_rate_pocket=("pressured_pocket", "mean"),
             avg_ttp_pocket=("time_to_pressure_pocket_sec", "mean"),
             mean_epa=("epa", "mean"),
             mean_def_effect=("def_effect", "mean"),
         )
         .sort_values("mean_def_effect", ascending=False)
)

print(cluster_summary)


# =========================
# 16) VISUAL: mean trajectory per cluster
# =========================
def plot_cluster_mean(cluster_id):
    idx = meta["cluster_abs"].eq(cluster_id).to_numpy()
    Xc = X[idx]
    if len(Xc) == 0:
        print("No samples for cluster", cluster_id)
        return

    x = Xc[:, :K_FRAMES].mean(axis=0)
    y = Xc[:, K_FRAMES:].mean(axis=0)

    label = cluster_names.get(cluster_id, f"Cluster {cluster_id}")

    plt.figure()
    plt.plot(x, y, marker="o")
    plt.title(f"{label}: mean pass rush trajectory (ball at origin)")
    plt.xlabel("x_rel (yards)")
    plt.ylabel("y_rel (yards)")
    plt.show()

for c in range(CLUSTERS_K):
    plot_cluster_mean(c)


# =========================
# 17) SAVE OUTPUTS
# =========================
meta2.to_csv("rusher_clusters_meta2.csv", index=False)
np.save("rusher_clusters_X.npy", X)

print("Saved: rusher_clusters_meta2.csv and rusher_clusters_X.npy")
```
